```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var=sns.load_dataset("tips")
var
```

```python
sns.violinplot(x="day",y="total_bill",data=var,hue="time")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,linewidth=2,palette="Dark2")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,order=["Dinner","Lunch"])
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,linewidth=2,palette="Dark2",saturation=0-4)
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,linewidth=2,color="y")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,scale="width")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,inner="point")
plt.show()
```

```python
sns.violinplot(y=var["total_bill"])
plt.show()
```
