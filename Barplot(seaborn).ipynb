```python
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd
```

```python
var=sns.load_dataset("penguins")
var
```

```python
sns.barplot(x=var.island,y=var.bill_depth_mm)
plt.show()
```

```python
sns.set(style="dark")
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_depth_mm",data=var,hue="sex",n_boot=5,saturation=20,errcolor="y",dodge=True,alpha=0.9)
plt.show()
```
