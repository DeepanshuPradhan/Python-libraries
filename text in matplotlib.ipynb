```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[3,1,5,2,6]

plt.plot(x,y)
plt.title("DEEPANSHU",fontsize=20)
plt.text(2,5,"python",fontsize=10,style="italic",bbox={"facecolor":"red"})
plt.annotate("python",xy=(3,2),xytext=(4,4),arrowprops=dict(facecolor="black",shrink=10))
plt.legend(["up"],facecolor="red",edgecolor="yellow",framealpha=0.5,shadow=True)
plt.show()
```
