```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var=sns.load_dataset("tips")
var
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,hue="time",color="r",order=["Fri","Sun","Thur","Sat"], showmeans=True)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day", y="total_bill", data=var, color="r", showmeans=True, meanprops={"marker": "+", "markeredgecolor": "b"})
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot( data=var, color="r", showmeans=True,orient="h" )
plt.show()
```

```python
sns.boxplot( x=var["total_bill"], color="r", showmeans=True )
plt.show()
```
