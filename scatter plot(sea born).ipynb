```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var=sns.load_dataset("penguins").head(30)
var
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,hue="sex")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,hue="sex",style="sex",size="sex")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,hue="sex",style="sex",size="sex",sizes=(10,40))
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm", data=var,hue="sex",palette="gist_ncar",sizes=(10,40))
plt.show()
```

```python
m={"Male":"*","Female":"o"}
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm", data=var,hue="sex",palette="gist_ncar",sizes=(10,40),alpha=0.5,markers="m")
plt.show()
```

```python
m={"Male":"o","Female":"*"}
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm", data=var,alpha=0.5,style="sex",markers=m)
plt.show()
```

```python
m={"Male":"*", "Female":"o"}
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=var, hue="sex", palette="gist_ncar", sizes=(10,40), alpha=0.5, style="sex",markers=m)
plt.show()
```
