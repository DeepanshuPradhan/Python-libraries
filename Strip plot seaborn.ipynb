```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var=sns.load_dataset("tips")
var
```

```python
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r",linewidth=2,edgecolor="r")
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r",alpha=0.5)
plt.show()
```

```python
sns.stripplot(x=var["total_bill"])
plt.show()
```
