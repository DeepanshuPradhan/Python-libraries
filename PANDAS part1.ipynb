```python
import pandas as pd

# Convert list_1 to DataFrame and transpose
list_1 = ([1, 2, 3, 4, 5], [11, 12, 13, 14, 15])
df = pd.DataFrame(list_1).T

# Name the columns
df.columns = ['A', 'B']

# Perform conditional operations
df['python'] = df['A'] <= 3
df['python_1'] = df['B'] >= 12

# Print the resulting DataFrame
(df)
```
