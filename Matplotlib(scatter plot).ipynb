```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[2,3,1,4,5,3]
plt.scatter(x,y)
plt.title("Scatter Plot",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6]
no=[2,3,1,4,5,3]
no2=[3,2,4,6,7,6]
colors=[10,49,30,29,56,20]
sizes=[100,270,400,200,25,100,]
plt.scatter(day,no,c=colors,s=sizes,cmap="Accent_r")
plt.scatter(day,no2,c="b",s=sizes,cmap="greens")
t=plt.colorbar()

plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day")
plt.ylabel("No.")
plt.show()
```
