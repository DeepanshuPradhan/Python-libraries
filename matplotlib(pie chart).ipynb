```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["c","C++","java","python"]


plt.pie(x,labels=y,colors=["r","b","g","y"],textprops={"fontsize":15},
        wedgeprops={"linewidth":4},center=(2,2),rotatelabels=True)
plt.title("DEEPANSHU")
plt.legend(loc=2)
plt.show()

```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
x1=[40,30,20,10]
y=["c","C++","java","python"]

c=["r","b","g","y"]

plt.pie(x,labels=y,radius=1)
plt.pie(x1,radius=0.5,colors=c)
plt.title("DEEPANSHU")

plt.show()

```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
x1=[40,30,20,10]
y=["c","C++","java","python"]

c=["r","b","g","y"]

plt.pie(x,labels=y,radius=1.5,colors=c)
cr=plt.Circle(xy=(0,0),radius=1,facecolor="w")
plt.gca().add_artist(cr)
plt.show()
```
